Create Dataframe

What is dataframe?
=> A DataFrame is a two-dimensional, labeled data structure in Pandas that stores data in rows and columns, similar to a table. Each column can hold different data types, and it provides powerful tools for data manipulation, analysis, and cleaning

<br>

Think of it as an Excel sheet inside Python, but:

```python
print(np.arange(0,20).reshape(5,4))         # arange is used to create a sequence of numbers, and reshape is used to change the shape of the array. In this case, it creates a 5x4 array with numbers from 0 to 19.
```

```python
# Create DataFrame
df = pd.DataFrame(data = np.arange(0,20).reshape(5,4), index = ["Row1","Row2","Row3","Row4","Row5"], columns = ["Column1","Column2","Column3","Column4"])

print(df)

df.head()               # by default display first 5 rows
df.tail()               # by default display last 5 rows
# df.head(2)

print(type(df))
```

```python
df.describe()
```

# Indexing 2 vabe kora jay Pandas e
1. column name diye
2. loc and iloc method diye rowindex[loc] and rownumber[iloc]         location
3. row index number or column index number diye   [iloc ] diye

1. Indexing using Column name:

```python
# Using Column name

#df["Column1"]       # single column
df[["Column1","Column2","Column3"]]


# series vs DataFrame
print(type(df["Column1"]))      # series
print(type(df[["Column1","Column2"]]))    # DataFrame
```

2. Indexing using rowindex name (loc method): 

```python
# Using row index name ( loc diye)

print(type(df.loc["Row2"]))         # Series
df.loc["Row2"]

print(type(df.loc[["Row2"]]))       # DataFrame
df.loc[["Row2","Row3"]]
```

3. Indexing using both column index number and row index number (iloc method):

```python
# df.iloc[row_index_number(starting index: ending index+1), column_index_number(starting index: ending index+1)]
df.iloc[2:4,0:2]
```

Convert DataFrame to arrays

```python
df.iloc[:,:].values     # convert DataFrame to array
```

Basic Operations


```python
df = pd.DataFrame(data = np.array([[0,1,2,3],[4,5,6,np.nan],[8,np.nan,10,11],[12,13,14,15],[16,np.nan,18,19]]), index = ["Row1","Row2","Row3","Row4","Row5"], columns = ["Column1","Column2","Column3","Column4"])
print(df)
print("\n")
print(df.isnull())   # check for null values
print("\n")
print(df.isnull().sum())      # count of null values in each column
```

```python
df["Column2"].value_counts()    # count of unique values in a column
```

```python
df["Column2"].unique()          # unique values in a column
```

```python
# df["Column2"] > 2
df > 4
# df[df > 4]        # this will give all the values greater than 4 and rest will be NaN
# df[df["Column2"] > 2]     this will give all the rows where Column2 value is greater than 2
# df[df["Column2"] > 2][["Column1","Column3"]]     this will give only Column1 and Column3 values where Column2 value is greater than 2
```
