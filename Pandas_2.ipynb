# Pandas Read CSV Files, StringIO

```python
from io import StringIO

import pandas as pd
import numpy as np

df = pd.read_csv("Formula.csv")         # Reading CSV file ( converts CSV to DataFrame )
print(type(df))

print(df)
print("\n")
```

String -> CSV File -> DataFrame

```python
# convert string to in memory file format object

data = ("Col1,Col2,Col3\n"
        "x,y,1\n"
        "a,b,2\n"
        "c,d,3\n")



string_io_data = StringIO(data)
print(type(string_io_data))
df2 = pd.read_csv(string_io_data,sep = ",")  # Reading CSV from in memory file object
print(df2)
print("\n")


# or directly
#df3 = pd.read_csv(StringIO(data), sep = ",")
df3 = pd.read_csv(StringIO(data))           # sep = "," is default for CSV
print(df3)

```

```python
# usecols -> to read specific columns from CSV
pd.read_csv(StringIO(data), usecols = ["Col1","Col3"])
```

```python
df = pd.read_csv(StringIO(data), dtype = {"Col3": float})   # specify data type of a column
print(df)

df.info()    # get info about dataframe
```

```python
df.isnull()
```

```python
# Select element using indexing
print(df.iloc[1,2])   # 2nd row, 3rd column

print(df["Col3"][1])

print(df)
```

```python
data = ("Col1,Col2,Col3\n"
        "1,2,3\n"
        "4,5,6\n"
        "7,8,9\n")

df = pd.read_csv(StringIO(data), dtype = {"Col1": int, "Col2": float, "Col3": int})   # specify data type of a column
df.info()
```

Create CSV file from dataframe

```python
ndf = pd.read_csv(StringIO(data), usecols = ["Col1","Col3"])
print(ndf)
print("\n")

# Create CSV file
ndf.to_csv("NewFile.csv", index = False)   # index = False to avoid writing index column    

```

(Index_col): to treat a column as index

```python
data = ("Index,Col1,Col2,Col3\n"
        "24,1,2,3\n"
        "25,4,5,6\n"
        "26,7,8,9\n")
pd.read_csv(StringIO(data), index_col = 0)   # (Index_col): to treat a column as index


```
