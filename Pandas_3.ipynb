# Working With Python Pandas JSON

Website to check if valid json data or not :     JSONLint.com

```python
import pandas as pd
from io import StringIO
data = '{"Name":"Sadman Zaman","Email":"sadmanzaman123@gmail.com","Job_Profile":[{"Title1":"Data Scientist","Title2":"ML Engineer"}]}'
type(data)          # str

json_data = pd.read_json(StringIO(data))                # convert JSON string to pandas DataFrame
type(json_data)    # pandas.core.frame.DataFrame
print(json_data)
```

```python
json_data.to_json(orient = "records")   # convert pandas DataFrame to JSON string with 'records' orientation
json_data.to_json(orient = "index")    # convert pandas DataFrame to JSON string with 'index' orientation
```
