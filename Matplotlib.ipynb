# Univariate

```python
# univariate - Numerical
data = {
    "Salary" : [50000, 60000, 55000, 70000, 65000, 80000, 72000, 68000, 75000, 62000]
}

df = pd.DataFrame(data)

print(df.shape)         # 10 ta data 1 ta column

df
```

```python
# Line Plot
plt.plot(df["Salary"], color = "blue", marker = "*", linestyle = ":", linewidth = "2")          # df[column_name]
plt.grid()
plt.show()
```

```python
# histogram
plt.hist(df["Salary"], bins = 5, color = "green", edgecolor = "black")  # bins = number of bars
plt.show()
```

```python
# Box Plot
df.loc[11] = 120000  # adding outlier
plt.boxplot(df["Salary"])
plt.show()

# Box Plot
df.loc[11] = 120000  # adding outlier
plt.boxplot(df["Salary"])
plt.show()

```

Univariale: Categorical

```python
df["dept"] = ["HR", "Admin","IT", "IT", "Finance", "HR", "IT", "Finance", "IT", "Finance", "IT"]
df
```

```python
# Pie Chart
count = df["dept"].value_counts()
count
plt.pie(count, labels = count.index, autopct = "%1.1f%%")
plt.show()
```

```python
plt.pie(count, labels = count.index, autopct="%1.1f", explode = [0,0.1,0.2,0.3])   # autopct="%1.1f" -> to show number   
# explode -> to separate the slices
plt.show()
```

```python
# Count Plot
plt.bar(count.index,count, color = ["black","red", "yellow"])   # default color black
plt.show()
```

# Bivariate 

```python
# Bivariate - numerical - numerical

df["age"] = [25,30,28,35,32,40,38,29,33,31,34]
df

```

```python
# Scatter plot
plt.scatter(df["age"], df["Salary"], color = "orange")  # x = age, y = Salary, default color blue

#extras
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("Age vs Salary")

plt.show()
```

```python
sort_age = df.sort_values("age")
```

```python
# Line Plot

plt.plot(sort_age["age"], df["Salary"], color = "red", marker = "o", linewidth = "2")
plt.grid()
plt.show()
```

```python
# Bar Chart
plt.bar(sort_age["age"], df["Salary"], color = "green", align = "edge")
plt.show()
```

# Bivariate: Numerical & Categorical data

```python
hr_sal = df[df["dept"] == "HR"]["Salary"]
hr_sal
```

```python
it_sal = df[df["dept"] == "IT"]["Salary"]
it_sal
```

```python
fin_sal = df[df["dept"] == "Finance"]["Salary"]
fin_sal
```

```python
# using Box Plot
plt.boxplot([hr_sal, it_sal, fin_sal], tick_labels= ["HR","IT", "Finance"])     # age eta chilo (labels = )
plt.legend()
plt.show()
```

```python
# using pie chart

# salary_by_dept = df.groupby("dept")["Salary"].mean()          oi dept er sobar mean salary
salary_by_dept = df.groupby("dept")["Salary"].sum()             # oi dept er sobar salaryr sum / total salary
salary_by_dept

plt.pie(salary_by_dept, labels = salary_by_dept.index, autopct= "%1.2f", shadow = True, explode = [0.1, 0, 0.1])
plt.axis("equal")
plt.show()
```

```python
# using bar plot

hr_mean = sum(hr_sal)/len(hr_sal)
it_mean = sum(it_sal)/len(it_sal)
fin_mean = sum(fin_sal)/len(fin_sal)

plt.bar(["HR","IT","Finance"], [hr_mean,it_mean,fin_mean], color = ["green", "black", "red"])
plt.show()      # avg salary of the 3 depts
```

# Multivariate Analysis

Multivariate Analysis : 3 Numericals

```python
df["experience"] = [1, 2, 4, 2.5, 1, 6, 5, 5, 1.5, 3]
df
```

```python
# Bubble Plot:

plt.scatter(df["age"], df["Salary"], s = df["experience"] * 15, color = "skyblue", edgecolor = "black")       
# s = df["experience"]*10  is the size of the bubble

plt.title("Age vs Salary vs Experience")
plt.xlabel("Age")
plt.ylabel("Salary")
#plt.grid()

plt.show()
```

Multivariate Analysis : 2 Numericals & 1 Categorical Column

```python
plt.scatter(df["age"], df["Salary"], c = df["dept"].map({"HR":"yellow", "IT":"blue", "Finance":"orange"}))
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("Age vs Salary vs Dept")
#plt.legend()                       cannot use this one
plt.show()
```

```python
# use legend

color = {"HR": "yellow", "IT": "blue", "Finance": "orange"}

for dept, color in color.items():
    df_dept = df[df["dept"] == dept]
    plt.scatter(df_dept["age"], df_dept["Salary"], c = color, label = dept)

plt.legend()
```
